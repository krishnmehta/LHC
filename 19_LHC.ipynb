**Question 3**

```python
top_tag=[]
unique_tags = dict()
for x in data_long['tags']:
  top_tag.append(len(x.split('>'))-1)
  t=x.split('>')
  t.pop()
  for y in t:
    y=y.lower()
    if (y[1:] in unique_tags):
      unique_tags[y[1:]] += 1
    else:
      unique_tags[y[1:]] = 1
print(top_tag)
data_long['count']=top_tag
print(unique_tags)
```

```python
#Now lets sort the unique tag
sort1 = sorted(unique_tags, key=unique_tags.get, reverse=True)
print(sort1)
```

```python
#Now for top 25 tags
count=0
print("Top 25 tags are: ")
for x in sort1:
  print(x)
  count+=1
  if count==25:
    break
```
